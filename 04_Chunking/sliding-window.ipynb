### #Sliding Window Chunking

```python
from pypdf import PdfReader
from nltk import sent_tokenize
```

```python
pdf_path = "./docs/PAPER1.pdf"
reader = PdfReader(pdf_path)
text = ""
for page in reader.pages:
    text += page.extract_text() + "\n"

print(f"PDF loaded. Total characters: {len(text)}")
```

```python
sentences = sent_tokenize(text)
```

```python
window_size = 5  # sentences per window
overlap = 2
sliding_chunks = []
for i in range(0, len(sentences), window_size - overlap):
    sliding_chunks.append(" ".join(sentences[i:i+window_size]))
print(f"Sliding window chunks: {len(sliding_chunks)}")
```

```python
sliding_chunks[5]
```

```python
from langchain_ollama import OllamaEmbeddings

ollama_embed = OllamaEmbeddings(model="mxbai-embed-large:latest")
embedded_docs = ollama_embed.embed_documents(sliding_chunks)

embedded_docs[0][:10]
```
